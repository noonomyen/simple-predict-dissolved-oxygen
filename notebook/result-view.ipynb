# Result View

```python
DATA_PATH = "../output/data"
ALGO = ["mlr", "mpr", "lstm", "gru"]
```

```python
from pathlib import Path
from pandas import DataFrame, read_csv
from IPython.display import display
```

```python
DATA_PATH = Path(DATA_PATH).resolve()
```

```python
comps = {
    "algorithm": [],
    "avg_mae": [],
    "avg_mse": [],
    "avg_rmse": [],
    "avg_r2": []
}

for algo in ALGO:
    data = read_csv(DATA_PATH / algo / "results.csv")
    
    comps["algorithm"].append(algo.upper())
    comps["avg_mae"].append(data["mae"].mean())
    comps["avg_mse"].append(data["mse"].mean())
    comps["avg_rmse"].append(data["rmse"].mean())
    comps["avg_r2"].append(data["r2"].mean())

    print(algo.upper())
    data.sort_values("mse", inplace=True)
    display(data.round(3))
    print()
```

```python
comps = DataFrame(comps)
comps.sort_values("avg_mse", inplace=True)
display(comps.round(3))
comps.to_csv(DATA_PATH / "result-comparison.csv", index=False)
```
